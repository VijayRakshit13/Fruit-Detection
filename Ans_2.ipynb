```python
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

Finding the feature vector of the training dataset

```python
fruits=os.listdir(path)
feature=dict()
c=0
for fruit in fruits:
  fd=[]
  npath="/content/Fruit-Images-Dataset/Training/"+fruit
  nimages=os.listdir(npath)
  for x in nimages:
    c=c+1
    
    img=imread(npath+"/"+x)
    resized_img = resize(img, (128,64))
    fd= hog(resized_img, orientations=8, pixels_per_cell=(8, 8), cells_per_block=(1, 1), visualize=False, multichannel=True)
    if c<10:
      feature[fruit+"0000"+str(c)]=fd
    elif c>=10 and c<100:
      feature[fruit+"000"+str(c)]=fd
    elif c>=100 and c<1000:
      feature[fruit+"00"+str(c)]=fd
    elif c>=1000 and c<10000:
      feature[fruit+"0"+str(c)]=fd
    else:
      feature[fruit+str(c)]=fd
```

Finding the feature vector of the test images and then
Finding the norm between the feature vectors of training data and test data
and then printing the test image label 

```python
tpath ="/content/Fruit-Images-Dataset/Test"
tfruits=os.listdir(tpath)
c=0
t=0
su=0
for fruit in tfruits:
  tspath="/content/Fruit-Images-Dataset/Test/"+fruit
  timages=os.listdir(tspath)
  for x in timages:
    dist=dict()
    c=c+1
    
    knn=[]
    img=imread(tspath+"/"+x)
    resized_img = resize(img, (128,64))
    tfd= hog(resized_img, orientations=8, pixels_per_cell=(8, 8), cells_per_block=(1, 1), visualize=False, multichannel=True)
    
    for i in feature.keys():
      dist[i]= np.linalg.norm(feature[i] - tfd)
    
    sortdist=dict(sorted(dist.items(), key=lambda item: item[1]))
    cnt=0
    k=3
    for i in sortdist.keys():
      if cnt<k:
        knn.append(i[:-5])
      cnt=cnt+1
    max=0
    for i in knn:
      y=knn.count(i)
      if(y>max):
        y=max
        categ=i

    
    #accuracy
    if categ==fruit:
      su=su+1
      print(categ)
    else:
      print(categ+" "+fruit)
    t=t+1
print(str(su)+" "+str(t))
print("accuracy="+str((su/t)*100))
```
